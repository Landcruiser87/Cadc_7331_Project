# The Wonderful World of Coffee
### Business Understanding
This notebook explores a coffee dataset that has been scraped by a reddit user from the Coffee Quality
Institute's website and uploaded onto GitHub.  Our idea behind researching this dataset is to identify where the top
coffee brands come from, and what attributes go into the production of that coffee that makes it so desirable?
Can a model be built targeting those ranges of successful coffee producers in order to predict ratings for 
their future brands?  These are some of the questions we will investigate in our first project.

The data source for our dataset:

https://github.com/jldbc/coffee-quality-database

### Data Description (Meaning/Type/Quality)
Lets import our libraries and data. 


```python
#Add library references
import pandas as pd
import numpy as np
import seaborn as sns
#import plotly.plotly as py
#import plotly.graph_objs as go
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Upload Data
df_ar = pd.read_csv('https://raw.githubusercontent.com/jldbc/coffee-quality-database/master/data/arabica_data_cleaned.csv',
                    sep=',', header=0) # read in the arabicaica data
df_rob = pd.read_csv('https://raw.githubusercontent.com/jldbc/coffee-quality-database/master/data/robusta_data_cleaned.csv',
                     sep=',', header=0) # read in the Robusta data
#Column rename to match for merging
df_ar.rename(columns={'Unnamed: 0':'Id'}, inplace=True)
df_rob.rename(columns={'Unnamed: 0':'Id',
                       'Bitter...Sweet':'Sweetness',
                       'Uniform.Cup':'Uniformity',
                       'Salt...Acid':'Acidity',
                       'Fragrance...Aroma':'Aroma'}, inplace=True)
```

Since the data came to us in two CSV's of arabica and robusta, lets combine the two datasets to begin our analysis.  
First we'll need to remove a few columns and merge the two dataframes.

```python

#dropping columns we won't use
df_rob = df_rob.drop(['Lot.Number', 'altitude_low_meters', 'altitude_high_meters', 'Certification.Contact',
                      'Certification.Contact', 'Expiration', 'Certification.Body', 'ICO.Number',
                      'Certification.Address','Mouthfeel', 'Id'], axis=1)
df_ar = df_ar.drop(['Lot.Number', 'altitude_low_meters', 'altitude_high_meters', 'Certification.Contact',
                      'Certification.Contact', 'Expiration', 'Certification.Body', 'ICO.Number',
                      'Certification.Address', 'Id'], axis=1)

df_comb = df_ar.append(df_rob)
```

### Data Quality & Simple Statistics
Now that our dataframes are combined, lets analyze counts of missing values and simple statistics.
#### Missing Values

```python
print("Structure of data:\n",df_comb.shape,"\n")
print("Count of missing values:\n",df_comb.isnull().sum().sort_values(ascending=False),"\n")
```

The majority of missing values center around farm name, mill, producer, altitude, company.  At this stage, we need
to decide what categorical values we can keep for our analysis.  We believe that alot of those categorical missing values 
are because the companies surveyed didn't have the available information or didn't fill out their survey completely.
Country/region might be one of the better attributes to start an analysis due to its low missing value count. 
The continuous data has very low NA counts which means any regression, will likely fair well.


One area we think influences coffee production is altitude_mean_meters.  To clean that column, we'll be be 
replacing blank and altitude means of 1 because we believe them to be a data entry error.  We also removed a few rows
in Origin, Clean.Cup, and Body that we classified as erroneous.


```python
#Changing datatypes
conv_dict = {'Species': str,
                'Owner': str,
                'Mill': str,
                'Company': str,
                'Region': str,
                'Producer': str,
                'Variety': str
                }
df_comb = df_comb.astype(conv_dict)

#Outlier Removal altitude
df_comb = df_comb.replace({'altitude_mean_meters': {1: df_comb['altitude_mean_meters'].mean()}})
df_comb.loc[[896,1040,1144,543],'altitude_mean_meters'] = np.nan
df_comb['altitude_mean_meters'].fillna((df_comb['altitude_mean_meters'].mean()),inplace=True)
df_comb = df_comb[df_comb["Body"]>1]  #remove row with 0 ratings 
#nan removal from country
df_comb = df_comb.drop(df_comb.index[1197])
#removal of Clean.Cup Outlier
df_comb = df_comb[df_comb['Clean.Cup']>0] 
#second check of missing values to ensure integrity. 
print("Count of missing values:\n",df_comb.isnull().sum().sort_values(ascending=False),"\n")

```

### Visualize Attributes
Now that we've got our data organized and a little cleaner, lets begin visualizing our data. We'll start with a 
bar chart of the Number of coffee samples by Country.  


```python
counts = df_comb['Country.of.Origin'].value_counts().to_dict()
min_count = min(counts.values())
max_count = max(counts.values())

#Bar graph of number of coffee samples per country, top 50
counts_top50 = dict(list(counts.items())[int(len(counts)/2):])
plt.figure(figsize=(20,10))
plt.bar(range(len(counts)), list(counts.values()), align='center')
plt.xticks(range(len(counts)), list(counts.keys()), rotation=90)
plt.title("Number of Coffee Samples by Country", fontsize='18')
```

coming in first, we're not surprised by those that follow it.  Columbia, Guatemala, Brazil, Taiwan are all well known

Now that we've got an idea behind who are the biggest producers, lets see what countries produce the highest rated 
coffee via the reviewers "Total Cup Score."  We'll group by country and analyze their average score.

```python
#Dataframe of countries by avg Cup rating

country_lists=list(df_comb['Country.of.Origin'].unique())
country_total_cup_ratio=[]
for each in country_lists:
    country=df_comb[df_comb['Country.of.Origin']==each]
    country_total_cup_avg=round(sum(country['Total.Cup.Points'])/len(country),2)
    country_total_cup_ratio.append(country_total_cup_avg)

    
data=pd.DataFrame({'Country of Origin':country_lists,'Total Cup Avg':country_total_cup_ratio})
new_index=(data['Total Cup Avg'].sort_values(ascending=False)).index.values
sorted_data = data.reindex(new_index)
sorted_data.head(10)
```

```python
import matplotlib.colors
plt.figure(figsize = (20,10))
cmap = plt.cm.coolwarm_r
norm = matplotlib.colors.Normalize(vmin = 1, vmax = 30)
plt.bar(height=sorted_data['Total Cup Avg']-75,
        x = sorted_data['Country of Origin'],
        align = 'center',bottom = 75,
        color = cmap(norm(sorted_data['Total Cup Avg'].values-75)))
plt.xticks(rotation=90)
plt.title("Total coffee score by country", fontsize='18')
```

Here we see that quantity does not equal quality as many of the top producing Central American countries
are lower in terms of overall coffee rating.  It seems their business model focus's more on higher 
production than on quality.  Which makes sense as coffee's worldwide popularity grows and the more experienced 
manufacturing countries adjust to meet demand. The best rated coffee would comes from United States, 
Papua New Guinea, Ethiopia, and Japan.  


Now that we've explored some of the country categories, lets delve into the continuous variables themselves to get an 
idea behind their distributions.  Our next plot is a frequency plot of the continuous variables we initially predict to 
be influential.  Those are Aroma, Aftertaste, Balance, Flavor, Acidity, Moisture, Cupper.Points, and Total.Cup.Points.


```python
#Frequency plots 
col_names = ['Aroma','Aftertaste', 'Aroma','Balance', 
             'Flavor', 'Acidity','Moisture', 'Cupper.Points', 'Total.Cup.Points']

fig, ax = plt.subplots(len(col_names), figsize=(16,12))

for i, col_val in enumerate(col_names):

    sns.distplot(df_comb[col_val], hist=True, ax=ax[i])
    ax[i].set_title('Frequency distribution '+col_val, fontsize=10)
    ax[i].set_xlabel(col_val, fontsize=8)
    ax[i].set_ylabel('Count', fontsize=8)

plt.show()
```

These frequency plots show a good distribution of value's with some slight left skew.  There is an odd gap in a 
few of them.  Moisture has some interesting distributions with most centered around the mean or on the dry side of the
scale.  Which leads us to believe that people might prefer a dryer taste when it comes to rating their coffee.  


```python
#Histogram charts
df_num = df_comb.select_dtypes(include=['float64'])
df_num.hist(figsize =(14,12))
```


Building off the frequency plot, we generated histograms to see binning effects and look for any other traits.
Distributions look to be well centered for most numerical categories.  The Quaker category is very right skewed, 
but that's because its a measure of the quality of bean.  So a lower number actually bodes better for a coffee beans 
review.  Yet, for Clean.Cup, Uniformity, and Sweetness we see a heavy left skew.  Clean.Cup is a 
measure of defects in the flavor of the coffee, so we're not surprised that producers would only submit those beans 
they felt had the fewest defects.  Uniformity is the consistency from cup to cup, so that skew is also unsurprising.  
Most manufacturers would pick similar beans/fields to assure consistency in this category. Lastly, Sweetness is a 
description of the sugary intensity that the coffee exhibits.  This is more interesting as classic understanding of 
coffee is that bitter is the norm, yet for cuppers (those that taste coffee professionally)
a sweeter bean is apparently preferred.  

With Sweetness being a cupper preference, Does Acidity across different countries follow the same trend? Lets 
load up the acidity attribute and take a look.

```python
#Country Plotting Function by D.Josephs
def countrysorter(frame = df_comb,col='Total.Cup.Points'):
    contlist = list(frame['Country.of.Origin'].unique())
    contval = []
    for each in contlist:
        cont = frame[frame['Country.of.Origin']==each]
        contavg =  round(sum(cont[col])/len(cont),2)
        contval.append(contavg)


    dat = pd.DataFrame({'Country.of.Origin':contlist,'avgval':contval})
    nindex = (dat['avgval'].sort_values(ascending = False)).index.values
    sordat = dat.reindex(nindex)
    return(sordat)

def countryplotter(dat, sub = 0,
                   nmin = 1, nmax = 30,tit = 'TODO'):
    plt.figure(figsize = (20,10))
    cmap = plt.cm.coolwarm_r
    norm = matplotlib.colors.Normalize(vmin = nmin, vmax = nmax)
    # print(norm(dat['avgval'].values-nmin))
    plt.bar(height = dat['avgval']-sub,
            x = dat['Country.of.Origin'],
            align = 'center',
            bottom = sub,
            color = cmap(norm(dat['avgval'].values-nmin)))
    plt.xticks(rotation = 90)
    plt.title(tit,fontsize = 18)

countryplotter(dat = countrysorter(col = 'Acidity'),sub = 6,nmin = 3, nmax = 10, tit = 'Acidity by Country')
```

Interestingly, those countries that we found previously to be successful in a Total Rating, also come with some of the 
highest acidity ratings. Papua New Guinea, United States, Ethiopia and Kenya all had Top 5 rankings for Total Cup
Score, but also had a Top 5 Acidity Rating.  So we now know that Sweetness and Acidity both influence 
being a top rated country.  Japan was also rated in the top 5 of Total Cup Ratings, but strangely, they have an Acidity
rating thats much lower than the other top rated coffee's.


### Explore Joint Attributes

Now that we've looked at alot of the individual attributes lets start exploring the relationships between the data.  
A correlation heatmap is a great place to start.  


```python
#Generate Correlation HeatMap
colormap = sns.diverging_palette(220, 10, as_cmap=True)
f, ax = plt.subplots(figsize=(10, 10))
corr = df_comb.corr()

sns.heatmap(corr, cmap="coolwarm", annot=True, fmt=".1f",
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

We see some of our correlation suspicions from earlier confirmed.  The features with the highest correlation (>0.8) are:
* Flavor - Aftertaste
* Aftertaste - Total.Cup.Points
* Flavor - Total.Cup.Points
* Balance - Aftertaste
* Balance - Total.Cup.Points
* Cupper.Points - Total.Cup.Points

The cupper.points - Total.Cup.Points makes sense as the two rating scales are the final evaluations for the coffee.  Most of the other correlations are Total.Cup.Points with Flavor, Aftertaste, and Balance.  This leads us to believe those 3 tasting aspects would be the most challenging to discern differences.  Honestly, with the subtle differences in each of the tasting attributes, I'm surprised that correlations aren't higher, but we're dealing with professional coffee tasters so their pallettes are far more trained.

```python
# Pairplot matrix.  
sns.pairplot(df_comb, vars = ['Acidity','Aftertaste','Aroma','Balance',
             'Flavor', 'Moisture', 'Cupper.Points', 'Total.Cup.Points'],
             hue = 'Species');
```

In the figure above we have a plot showing all numeric features as a scatterplot vs all other numeric features. 
It is most important to view the bottom row, as this shows how each feature relates to total cup points, or overall 
coffee quality. We can see that acidity and aroma seem to have a strongly linear relationship with the total score, 
while flavor and balance seem to be much closer to a flat, zero relationship with score. As we can see, moisture does 
not tend to matter, however exceptionally moist coffees seem to have a good chance to score poorly.  
It is interesting to note that there appear to be two distinct groups of moisture level, while every other variable 
appears to be more or less continuous. In the figure below, we have narrowed down the amount of plots, and colored 
them by bean processing method in order to see how the methods in which the beans were cleaned affects the coffee

```python
g = sns.PairGrid(df_comb,vars=['Flavor',"Aftertaste",
                               "Body","Clean.Cup", "Aroma","Acidity"],hue="Processing.Method",palette = "muted")
g.map(plt.scatter, alpha=0.8)
g.add_legend();
```

Even after narrowing down the number of features, it is difficult to ascertain whether or not the technique used to wash/process the beans has much of an effect on the quality of the coffe produced. Let us examine this further by looking at a table comparing coffee quality (total score) and the bean processing technique.

```python
df_comb.groupby(['Processing.Method'])['Total.Cup.Points'].describe()
```

It would appear that the groups of processing methods have more or less the same mean. Let us affirm our beliefs with a violin plot:

```python
def violinplotter(cat,cont="Total.Cup.Points",dat=df_comb):
    fig, ax  = plt.subplots()
    fig.set_size_inches(14,14)
    ax = (sns
        .violinplot(x = cat,
            y = cont,
            data = dat,split=True))
    
```

```python
    
violinplotter("Processing.Method")
```

It is very clear from the preceding plots and table that processing method has little to no effect on the overall quality of the cup. Let us now go one step further back into the lifetime of a coffee bean, and see how its color relates to its overall quality:

```python
violinplotter("Color")
```

We can see there is some slight effect, but nothing two noticeable. We would need to collect more data, as there is no such thing as no color... Let us now look at how altitude, the last bean source related parameter that is (in practice) worthy of analysis:

```python
#assigning elevation categories
conditions = [
    (df_comb['altitude_mean_meters'] < 1000),
    (df_comb['altitude_mean_meters'] < 2000),
    (df_comb['altitude_mean_meters'] < 3000)]
choices = ['Low', 'Mid', 'High']
df_comb['alt_category'] = np.select(conditions, choices, default='High')
```

```python
violinplotter("alt_category")
```

This plot is not complete without a frequency comparison of each altitude range:

```python
df_comb.groupby('alt_category').agg(['count'])
```

This is a truly interesting result of our analysis. We see that even though high altitude beans are maybe less common, they have a chance to be better tasting than all the other beans, but are also much more likely to taste worse (based off distribution) Meanwhile, a mid altitude bean is likely to be consistent, all the mid altitude beans, even with 1111 of them, fit an extremely narrow distributio of score. It is difficult to tell, but there is a slight chance that at least within the coffees of this dataset the mid altitude coffee is likey to taste slightly better than high altitude coffee

```python
dotplot = (sns
 .FacetGrid(df_comb, hue='alt_category', height=10)
 .map(plt.scatter, 'Flavor', 'Total.Cup.Points')
 .add_legend()
 .set(
    title='Flavor vs Total Pts by altitude category',
    xlabel='Flavor',
    ylabel='Total Cup Points'
))
```

This plot tends to agree with the statements made in the previous paragraph; It seems that the high altitude beans tend to be "high risk, high reward", meaning they contain some of the best and worst coffees, in a very small sample size. Low altitude coffee seems to be consistently average, while the medium altitude, while slightly more spread out, seems to be clustered a bit higher on the flavor and quality scale than the others
